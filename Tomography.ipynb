This program can be used for calculate the graph of the density matrix of a two photon state.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_VV = 896
N_VH = 18
N_VD = 297
N_VA = 437
N_VR = 652
N_VL = 50

N_HV = 50
N_HH = 836
N_HD = 433
N_HA = 615
N_HR = 729
N_HL = 520

N_DV = 197
N_DH = 566
N_DD = 715
N_DA = 145
N_DR = 596
N_DL = 354

N_AV = 459
N_AH = 371
N_AD = 91
N_AA = 766
N_AR = 242
N_AL = 371

N_RV = 111
N_RH = 630
N_RD = 324
N_RA = 360
N_RR = 287
N_RL = 641

N_LV = 284
N_LH = 215
N_LD = 332
N_LA = 621
N_LR = 733
N_LL = 84
```

```python
P_VV = N_VV / (N_VV + N_HH + N_VH + N_HV)
P_HH = N_HH / (N_VV + N_HH + N_VH + N_HV)
P_VH = N_VH / (N_VV + N_HH + N_VH + N_HV)
P_HV = N_HV / (N_VV + N_HH + N_VH + N_HV)

P_VD = N_VD / (N_VD + N_HA + N_VA + N_HD)
P_HA = N_HA / (N_VD + N_HA + N_VA + N_HD)
P_VA = N_VA / (N_VD + N_HA + N_VA + N_HD)
P_HD = N_HD / (N_VD + N_HA + N_VA + N_HD)

P_VR = N_VR / (N_VR + N_HL + N_VL + N_HR)
P_HL = N_HL / (N_VR + N_HL + N_VL + N_HR)
P_VL = N_VL / (N_VR + N_HL + N_VL + N_HR)
P_HR = N_HR / (N_VR + N_HL + N_VL + N_HR)


P_DD = N_DD / (N_DD + N_AA + N_DA + N_AD)
P_AA = N_AA / (N_DD + N_AA + N_DA + N_AD)
P_DA = N_DA / (N_DD + N_AA + N_DA + N_AD)
P_AD = N_AD / (N_DD + N_AA + N_DA + N_AD)

P_DV = N_DV / (N_DV + N_AH + N_DH + N_AV)
P_AH = N_AH / (N_DV + N_AH + N_DH + N_AV)
P_DH = N_DH / (N_DV + N_AH + N_DH + N_AV)
P_AV = N_AV / (N_DV + N_AH + N_DH + N_AV)

P_DR = N_DR / (N_DR + N_AL + N_DL + N_AR)
P_AL = N_AL / (N_DR + N_AL + N_DL + N_AR)
P_DL = N_DL / (N_DR + N_AL + N_DL + N_AR)
P_AR = N_AR / (N_DR + N_AL + N_DL + N_AR)


P_RR = N_RR / (N_RR + N_LL + N_RL + N_LR)
P_LL = N_LL / (N_RR + N_LL + N_RL + N_LR)
P_RL = N_RL / (N_RR + N_LL + N_RL + N_LR)
P_LR = N_LR / (N_RR + N_LL + N_RL + N_LR)

P_RV = N_RV / (N_RV + N_LH + N_RH + N_LV)
P_LH = N_LH / (N_RV + N_LH + N_RH + N_LV)
P_RH = N_RH / (N_RV + N_LH + N_RH + N_LV)
P_LV = N_LV / (N_RV + N_LH + N_RH + N_LV)

P_RD = N_RD / (N_RD + N_LA + N_RA + N_LD)
P_LA = N_LA / (N_RD + N_LA + N_RA + N_LD)
P_RA = N_RA / (N_RD + N_LA + N_RA + N_LD)
P_LD = N_LD / (N_RD + N_LA + N_RA + N_LD)
```

```python
S_00 = P_HH + P_HV + P_VH + P_VV
S_01 = P_HD - P_HA + P_VD - P_VA
S_02 = P_HL - P_HR + P_VL - P_VR
S_03 = P_HH - P_HV + P_VH - P_VV
S_10 = P_DH + P_DV - P_AH - P_AV
S_11 = P_DD - P_DA - P_AD + P_AA
S_12 = P_DL - P_DR - P_AL + P_AR
S_13 = P_DH - P_DV - P_AH + P_AV
S_20 = P_LH + P_LV - P_RH - P_RV
S_21 = P_LD - P_LA - P_RD + P_RA
S_22 = P_LL - P_LR - P_RL + P_RR
S_23 = P_LH - P_LV - P_RH + P_RV
S_30 = P_HH + P_HV - P_VH - P_VV
S_31 = P_HD - P_HA - P_VD + P_VA
S_32 = P_HL - P_HR - P_VL + P_VR
S_33 = P_HH - P_HV - P_VH + P_VV

S = np.array([[S_00, S_01, S_02, S_03], [S_10, S_11, S_12, S_13], [S_20, S_21, S_22, S_23], [S_30, S_31, S_32, S_33]])
```

```python

sigma_0 = np.array([[1, 0], [0, 1]])
sigma_1 = np.array([[0, 1], [1, 0]])
sigma_2 = np.array([[0, -1j], [1j, 0]])
sigma_3 = np.array([[1, 0], [0, -1]])

sigma = np.array([sigma_0, sigma_1, sigma_2, sigma_3])
```

```python
rho = np.zeros([4,4])

for i in range(4):
    for j in range(4):
        rho = rho + (1/4) * S[i][j] * np.kron(sigma[i], sigma[j])


print(rho)
```

```python
from qiskit.visualization import plot_state_city
plot_state_city(rho, title='Density Matrix')
```

```python
# Extraer la parte real e imaginaria de la matriz
parte_real = np.real(rho)
parte_imaginaria = np.imag(rho)

# Crear una figura con dos subgrficos
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Grfica para la parte real
im_real = axs[0].imshow(parte_real, cmap='viridis', origin='upper')
axs[0].set_title('Parte Real')
axs[0].set_xticks(np.arange(len(parte_real)))
axs[0].set_yticks(np.arange(len(parte_real)))
axs[0].set_xticklabels(["HH", "HV", "VH", "VV"])
axs[0].set_yticklabels(["HH", "HV", "VH", "VV"])
plt.colorbar(im_real, ax=axs[0])

# Grfica para la parte imaginaria
im_imaginaria = axs[1].imshow(parte_imaginaria, cmap='viridis', origin='upper')
axs[1].set_title('Parte Imaginaria')
axs[1].set_xticks(np.arange(len(parte_imaginaria)))
axs[1].set_yticks(np.arange(len(parte_imaginaria)))
axs[1].set_xticklabels(["HH", "HV", "VH", "VV"])
axs[1].set_yticklabels(["HH", "HV", "VH", "VV"])
plt.colorbar(im_imaginaria, ax=axs[1])
# Aadir etiquetas de los valores en las barras
for i in range(4):
    for j in range(4):
        axs[0].text(j, i, f'{parte_real[i, j]:.2f}', ha='center', va='center', color='white')
        axs[1].text(j, i, f'{parte_imaginaria[i, j]:.2f}', ha='center', va='center', color='white')

# Mostrar las grficas
plt.tight_layout()
plt.show()
```
